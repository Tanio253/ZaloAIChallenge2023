```python
class modelEvaluation:
  def __init__(self, config):
    self.tokenizer = AutoTokenizer.from_pretrained(config.model_id, revision = config.revision)
    self.generation_config = GenerationConfig(do_samples = False,
                                        top_p = 1,
                                        max_new_tokens = 128,
                                        pad_token_id = self.tokenizer.eos_token_id,
                                        trust_remote_code = True
                                        )
    self.model = AutoModelForCausalLM.from_pretrained(config.model_id,
                                                quantization_config = config.quantization_config,
                                                device_map = 'cuda',
                                                revision = config.revision,
                                                trust_remote_code = True
                                                )
    self.config = config
  def batch_data(sefl, data, batch_size):
    n = len(data)//batch_size
    batch = []
    for i in range(n):
      start = i*batch_size
      end = (i+1)*batch_size
      ba = data[start:end]
      batch.append(ba)
    batch.append(data[n*batch_size:])
    return batch

  def extract_answer(self, answer):
    ans = answer.split('Câu trả lời:')[1].strip()
    extracted_ans = [char for char in ans if char in ['A', 'B', 'C', 'D']]
    return extracted_ans[0]

  def preprocess_prompt(self, prompt):
    system_prompt = self.config.system_prompt
    input = self.tokenizer(f"""{system_prompt} {prompt}""", return_tensors = 'pt').to('cuda')
    return input

  def load_data(self, data_path)-> List[str]:
    with open(data_path) as f:
      data = json.load(f)
    data = data['data']
    preprocessed_data = []
    for d in data:
      prefix = f"""{self.config.system_prompt}"""
      question =  f" Câu hỏi:{d['question']}"
      choices_list = d['choices'].copy()
      choices = "".join(choices_list)
      choices = f" Lựa chọn:{choices}"
      answer = f" Câu trả lời:"
      completed_prompt = question+choices+answer
      preprocessed_data.append(completed_prompt)
    return preprocessed_data

  def extract_gt_ans(self, data_path)-> List[str]:
    with open(data_path) as f:
      data = json.load(f)
    data = data['data']
    gt = [d['answer'] for d in data]
    return gt

  def evaluate(self, data_path, batch_size=2):
    data = self.load_data(data_path)
    gt_ans = self.extract_gt_ans(data_path)
    batch_d = self.batch_data(data, batch_size)
    gen_ans = []
    correct_ans = []
    for batch in batch_d:
      tokenized_batch = [self.preprocess_prompt(b) for b in batch]
      outputs = [self.model.generate(**tb) for tb in tokenized_batch]
      ans = [self.tokenizer.batch_decode(o, skip_special_tokens = True) for o in outputs]
      gen_ans.extend(ans)
    for idx, (generated_ans, gt_ans) in enumerate(zip(gen_ans, gt_ans)):
      generated_an, = generated_ans
      y_pred = self.extract_answer(generated_an)
      if y_pred != None:
        correct_ans.append(y_pred == gt_ans)
      else:
        correct_ans.append(False)
    acc = sum(correct_ans)/len(correct_ans)
    print(f"Accuracy: {acc: .4f}\n")
    print(f"Invalid answer: {len(correct_ans)- sum(correct_ans)}")
```
